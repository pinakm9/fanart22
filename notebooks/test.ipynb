```python
import numpy as np
import cv2
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d
import sys
from pathlib import Path
from os.path import dirname, abspath
project_dir = Path(dirname(abspath('')))
module_dir = str(project_dir)
sys.path.insert(0, module_dir + '/modules')
print(module_dir)
import mod
```

```python
people = ["akhil", "arnab", "joydeep", "manisha", "prashant"]
songs = ["plastic-love", "veldt", "industry-baby", "marble-machine", "honesty"]

for i, person in enumerate(people):
    img_path = r"../assets/{}.png".format(person)
    song_path = r"../assets/{}.wav".format(songs[i])
    thumb_path = r"../assets/{}-thumb.png".format(person)
    video_path = r"../assets/{}-video.mp4".format(person)
    print("working on {}".format(person))
    obj = mod.Mod(img_path)
    X0 = np.random.uniform(low=[0., 0.],  high=[obj.width-1, obj.height-1], size=(5000, 2))
    obj.get_colormap()
    obj.find_circles(X0)
    obj.plot_steady(saveas=thumb_path, area_factor=20.)
    wave = mod.Wave(song_path)
    obj.animate(wave, saveas=video_path, move_factor=25., n_frames=2400)
```
